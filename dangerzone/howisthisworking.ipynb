```python
from io import StringIO
from html.parser import HTMLParser
from re import sub

class MLStripper(HTMLParser):
    def __init__(self):
        super().__init__()
        self.reset()
        self.strict = False
        self.convert_charrefs= True
        self.text = StringIO()
    def handle_data(self, d):
        self.text.write(d)
    def get_data(self):
        return self.text.getvalue()

def strip_tags(html):
    s = MLStripper()
    s.feed(html)
    return s.get_data() #some code i stright up stole but its okkkkkkk
```

```python
k_labels = ['Fluff', 'Hurt_Comfort', 'Angst', 'AU_Modern',\
       'Mutual_Pining', 'Established_Relationship', 'Light_Angst',\
       'Love_Confessions', 'Pining', 'Fluff_Humor'] #n1 from here

candidate_labels = ['Fluff', 'Hurt and Comfort', 'Angst', 'Modern AU',\
                      'Mutual Pining', 'Established Relationship', 'Light Angst',\
                      'Love Confessions', 'Pining', 'Fluff and Humor'] #n2 from here
 

def err_repl(n1, n2):
    etablerepl = []
    for x in range(len(k)):
        sqerr = (float(k[n1][x]) - float(pred_df[n2][x])) ** 2
        etablerepl.append(sqerr)
    errtable[n1] = etablerepl

for x in range(len(k_labels)):
    err_repl(k_labels[x], candidate_labels[x])
```
